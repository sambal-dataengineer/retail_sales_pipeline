What cleaning will be needed for Silver?
----------------------------------------------------------------------------------------

Based on the inspection, Silver should do these things:

Required Silver cleaning steps

1. Standardize column types
- convert Transaction Date → datetime
- convert Discount Applied → boolean
- convert Quantity → integer where valid

2. Handle missing numeric values

Use business-safe rules:

- if Price Per Unit and Quantity exist but Total Spent is null
→ compute Total Spent = Price Per Unit * Quantity
- if Total Spent and Quantity exist but Price Per Unit is null
→ compute Price Per Unit = Total Spent / Quantity
- if Total Spent and Price Per Unit exist but Quantity is null
→ compute Quantity = Total Spent / Price Per Unit
- only if it gives a valid whole-number quantity

3. Handle missing item values

Possible strategy:

- keep null as "Unknown Item" for analytics friendliness
- preserve original missingness in a flag column if needed

4. Handle missing discount values

Best Silver approach:

- convert True/False properly
- keep missing as NULL first
- optionally add a derived label later such as Unknown

5. Validate totals

Recheck:

- Total Spent = Price Per Unit × Quantity

after cleaning

6. Add data-quality flags

This dataset is a good candidate for flags like:

- item_missing_flag
- discount_missing_flag
- imputed_price_flag
- imputed_quantity_flag
- imputed_total_flag
